# Visualize MAE Reconstructions (75% / 90% / 95% Masking)

Load saved reconstruction grids or run evaluation and plot comparison across mask ratios.

```python
import matplotlib.pyplot as plt
from pathlib import Path

results = Path("../experiments/results/reconstruction")
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
for ax, name in zip(axes, ["mask75", "mask90", "mask95"]):
    p = results / name / "reconstruction_grid.png"
    if p.exists():
        ax.imshow(plt.imread(p))
    ax.set_title(name)
    ax.axis("off")
plt.tight_layout()
plt.show()
```
