# Embedding Spectrum Analysis

Load eigenvalue spectra and effective rank from `experiments/results/spectrum/{run_name}/` and plot spectral decay, effective rank across mask ratios.

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

base = Path("../experiments/results/spectrum")
for run in ["mask75", "mask90", "mask95"]:
    p = base / run / "eigenvalues.npy"
    if p.exists():
        ev = np.load(p)
        plt.semilogy(ev, label=run)
plt.xlabel("Index")
plt.ylabel("Eigenvalue")
plt.legend()
plt.title("Eigenvalue spectrum")
plt.tight_layout()
plt.show()
```

```python
effective_ranks = {}
for run in ["mask75", "mask90", "mask95"]:
    p = base / run / "effective_rank.npy"
    if p.exists():
        effective_ranks[run] = float(np.load(p).item())
print("Effective rank:", effective_ranks)
if effective_ranks:
    plt.bar(effective_ranks.keys(), effective_ranks.values())
    plt.ylabel("Effective rank")
    plt.tight_layout()
    plt.show()
```
